## Feature Engineering

Total Number of Delinquencies: Sum of delinquent months over the past year.

Rationale: Frequent delinquencies may indicate higher risk.

```python
data['total_delinquencies'] = data[['delq_m_1', 'delq_m_2', 'delq_m_3', 'delq_m_4', 'delq_m_5',
                                'delq_m_6', 'delq_m_7', 'delq_m_8', 'delq_m_9', 'delq_m_10',
                                'delq_m_11', 'delq_m_12']].sum(axis=1)
```

Maximum Consecutive Delinquencies: Longest streak of consecutive delinquent months.

Rationale: Extended periods of delinquency can signal financial distress.

```python
def max_consecutive_delinquencies(row):
    max_streak = current_streak = 0
    for delq in row:
        if delq > 0:
            current_streak += 1
            max_streak = max(max_streak, current_streak)
        else:
            current_streak = 0
    return max_streak

data['max_consecutive_delinquencies'] = data[['delq_m_1', 'delq_m_2', 'delq_m_3', 'delq_m_4', 'delq_m_5',
                                          'delq_m_6', 'delq_m_7', 'delq_m_8', 'delq_m_9', 'delq_m_10',
                                          'delq_m_11', 'delq_m_12']].apply(max_consecutive_delinquencies, axis=1)
```

Months Since Last Delinquency: Number of months since the most recent delinquency.

Rationale: Recent delinquencies may be more predictive of default risk.

```python
def months_since_last_delinquency(row):
    for i, delq in enumerate(row[::-1], 1):
        if delq > 0:
            return i
    return 13  # Indicates no delinquencies in the past 12 months

data['months_since_last_delinquency'] = data[['delq_m_1', 'delq_m_2', 'delq_m_3', 'delq_m_4', 'delq_m_5',
                                          'delq_m_6', 'delq_m_7', 'delq_m_8', 'delq_m_9', 'delq_m_10',
                                          'delq_m_11', 'delq_m_12']].apply(months_since_last_delinquency, axis=1)
```

Delinquency Trend: Slope of delinquency occurrences over time.

Rationale: An increasing trend may indicate worsening financial health.

```python
from scipy.stats import linregress

def delinquency_trend(row):
    months = list(range(1, 13))
    slope, _, _, _, _ = linregress(months, row)
    return slope

data['delinquency_trend'] = data[['delq_m_1', 'delq_m_2', 'delq_m_3', 'delq_m_4', 'delq_m_5',
                              'delq_m_6', 'delq_m_7', 'delq_m_8', 'delq_m_9', 'delq_m_10',
                              'delq_m_11', 'delq_m_12']].apply(delinquency_trend, axis=1)
```

Delinquency Volatility: Standard deviation of delinquency status over 12 months.

Rationale: High variability may reflect inconsistent payment behavior.

```python
data['delinquency_volatility'] = data[['delq_m_1', 'delq_m_2', 'delq_m_3', 'delq_m_4', 'delq_m_5',
                                   'delq_m_6', 'delq_m_7', 'delq_m_8', 'delq_m_9', 'delq_m_10',
                                   'delq_m_11', 'delq_m_12']].std(axis=1)
```

Early vs. Late Delinquencies: Comparison between delinquencies in the first 6 months and the last 6 months.

Rationale: Shifts in behavior can be indicative of changing financial situations.

```python
data['early_delinquencies'] = data[['delq_m_7', 'delq_m_8', 'delq_m_9', 'delq_m_10', 'delq_m_11', 'delq_m_12']].sum(axis=1)
data['late_delinquencies'] = data[['delq_m_1', 'delq_m_2', 'delq_m_3', 'delq_m_4', 'delq_m_5', 'delq_m_6']].sum(axis=1)
data['delinquency_shift'] = data['late_delinquencies'] - data['early_delinquencies']
```

Delinquency Ratio: Ratio of delinquent months to total months.

Rationale: Provides a normalized measure of delinquency frequency.

```python
data['delinquency_ratio'] = data['total_delinquencies'] / 12
```

Average Delinquency: Calculates the average delinquency status over the past 12 months.

Rationale: Provides insight into the overall delinquency behavior of the customer.

```python
data['average_delinquency'] = data[['delq_m_1', 'delq_m_2', 'delq_m_3', 'delq_m_4', 'delq_m_5',
                                'delq_m_6', 'delq_m_7', 'delq_m_8', 'delq_m_9', 'delq_m_10',
                                'delq_m_11', 'delq_m_12']].mean(axis=1)
```

Weighted Delinquency Score: Assigns higher weights to more recent months to emphasize recent delinquency behavior.

Rationale: Recent delinquencies may be more indicative of current financial status.

```python
import numpy as np

weights = np.arange(1, 13)
data['weighted_delinquency_score'] = data[['delq_m_1', 'delq_m_2', 'delq_m_3', 'delq_m_4', 'delq_m_5',
                                       'delq_m_6', 'delq_m_7', 'delq_m_8', 'delq_m_9', 'delq_m_10',
                                       'delq_m_11', 'delq_m_12']].apply(lambda row: np.dot(row, weights), axis=1)
```

Delinquency Frequency in Quarters: Counts the number of delinquencies in each quarter of the year.

Rationale: Identifies periods within the year where the customer is more prone to delinquency.

```python
data['Q1_delinquencies'] = data[['delq_m_1', 'delq_m_2', 'delq_m_3']].sum(axis=1)
data['Q2_delinquencies'] = data[['delq_m_4', 'delq_m_5', 'delq_m_6']].sum(axis=1)
data['Q3_delinquencies'] = data[['delq_m_7', 'delq_m_8', 'delq_m_9']].sum(axis=1)
data['Q4_delinquencies'] = data[['delq_m_10', 'delq_m_11', 'delq_m_12']].sum(axis=1)
```

Seasonal Delinquency Pattern: Detects if there is a seasonal pattern in delinquencies, such as higher rates during certain months.

Rationale: Seasonal trends can provide insights into external factors affecting payment behavior.

```python
data['seasonal_delinquency_pattern'] = data[['delq_m_1', 'delq_m_2', 'delq_m_3', 'delq_m_4', 'delq_m_5',
                                         'delq_m_6', 'delq_m_7', 'delq_m_8', 'delq_m_9', 'delq_m_10',
                                         'delq_m_11', 'delq_m_12']].apply(lambda row: np.std([row[i] - row[i+3] for i in range(9)]), axis=1)
```

Delinquency Recovery Rate: Measures the rate at which a customer returns to non-delinquent status after a delinquency.

Rationale: Indicates the customer's ability to recover from financial setbacks.

```python
def delinquency_recovery_rate(row):
    recoveries = 0
    for i in range(1, len(row)):
        if row[i-1] > 0 and row[i] == 0:
            recoveries += 1
    return recoveries

data['delinquency_recovery_rate'] = data[['delq_m_1', 'delq_m_2', 'delq_m_3', 'delq_m_4', 'delq_m_5',
                                      'delq_m_6', 'delq_m_7', 'delq_m_8', 'delq_m_9', 'delq_m_10',
                                      'delq_m_11', 'delq_m_12']].apply(delinquency_recovery_rate, axis=1)
```

Delinquency Acceleration: Assesses the change in delinquency frequency over time, indicating whether delinquencies are becoming more frequent.

```python
def delinquency_acceleration(row):
    first_half = row[:6].sum()
    second_half = row[6:].sum()
    return second_half - first_half

data['delinquency_acceleration'] = data[['delq_m_1', 'delq_m_2', 'delq_m_3', 'delq_m_4', 'delq_m_5',
                                     'delq_m_6', 'delq_m_7', 'delq_m_8', 'delq_m_9', 'delq_m_10',
                                     'delq_m_11', 'delq_m_12']].apply(delinquency_acceleration, axis=1)
```

Delinquency Persistence: Calculates the proportion of months with delinquencies over the total number of months.

Rationale: High persistence indicates consistent payment issues.

```python
data['delinquency_persistence'] = data['total_delinquencies'] / 12
```

First Delinquency Occurrence: Identifies the month in which the first delinquency occurred.

Rationale: Early occurrences may be more predictive of long-term default risk.

```python
def first_delinquency(row):
    for i in range(1, 13):
        if row[f'delq_m_{i}'] > 0:
            return i
    return np.nan

data['first_delinquency_month'] = data.apply(first_delinquency, axis=1)
```

```python
def delinquency_severity(row):
    severity = 0
    for i in range(1, 13):
        if row[f'delq_m_{i}'] > 0:
            severity += (13 - i)  # More recent months contribute more to the severity
    return severity

data['delinquency_severity_index'] = data.apply(delinquency_severity, axis=1)
```

These additional features delve deeper into the temporal patterns and severity of delinquencies, providing a more nuanced understanding of customer behavior in the credit card business. Incorporating these features can enhance predictive models aimed at assessing credit risk and customer reliability.

```python
from sklearn.feature_selection import mutual_info_classif, chi2
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np

def assess_predictive_power(data, target_col, features):
    results = []

    # Mutual Information
    print("Calculating Mutual Information...")
    X = data[features]
    y = data[target_col]
    mi_scores = mutual_info_classif(X, y, discrete_features='auto', random_state=42)
    mi_scores = pd.Series(mi_scores, index=features).sort_values(ascending=False)

    # Correlation (Pearson for numeric, Spearman for ordinal)
    print("Calculating Correlation...")
    correlations = []
    for feature in features:
        if data[feature].dtype in [np.float64, np.int64]:
            correlation = data[feature].corr(data[target_col], method='pearson')
        else:
            correlation = data[feature].corr(data[target_col], method='spearman')
        correlations.append(correlation)

    # Random Forest Feature Importance
    print("Calculating Feature Importances with Random Forest...")
    rf = RandomForestClassifier(random_state=42, n_estimators=100)
    rf.fit(X, y)
    feature_importances = pd.Series(rf.feature_importances_, index=features).sort_values(ascending=False)

    # Combine Results
    for feature in features:
        results.append({
            'Feature': feature,
            'Mutual_Info': mi_scores.get(feature, 0),
            'Correlation': correlations[features.index(feature)],
            'RF_Importance': feature_importances.get(feature, 0)
        })

    # Return as DataFrame
    result_df = pd.DataFrame(results).sort_values(by='Mutual_Info', ascending=False)
    return result_df
```

```python
target_col = 'Target'
features = [
    'delq_m_1', 'delq_m_2', 'delq_m_3', 'delq_m_4', 'delq_m_5', 'delq_m_6',
    'delq_m_7', 'delq_m_8', 'delq_m_9', 'delq_m_10', 'delq_m_11', 'delq_m_12',
    'total_delinquencies', 'max_consecutive_delinquencies', 'months_since_last_delinquency',
    'delinquency_trend', 'delinquency_volatility', 'early_delinquencies', 'late_delinquencies',
    'delinquency_shift', 'delinquency_ratio', 'average_delinquency', 'weighted_delinquency_score',
    'Q1_delinquencies', 'Q2_delinquencies', 'Q3_delinquencies', 'Q4_delinquencies',
    'seasonal_delinquency_pattern', 'delinquency_recovery_rate', 'delinquency_acceleration',
    'delinquency_persistence', 'first_delinquency_month', 'delinquency_severity_index'
]

predictive_power_df = assess_predictive_power(data, target_col, features)
print(predictive_power_df)
```

## Model Building -- Without SMOTE

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python


# Function to load and preprocess the data
def load_and_preprocess_data(file_path, cols=None):
    data = pd.read_csv(file_path)
    if cols:
        data = data[cols]
    X = data.drop(columns=['Id', 'Target'])
    y = data['Target']
    return X, y

# Function to split data into train and test
def split_data(X, y, test_size=0.3, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test


# Function for model training and tuning using Grid Search or RandomizedSearchCV
def train_and_tune_model(X_train, y_train, model_type='logistic', use_random_search=False):
    # Set model and hyperparameters
    if model_type == 'logistic':
        model = LogisticRegression()
        param_grid = {'C': [0.1, 1, 10], 'solver': ['liblinear', 'lbfgs']}
    elif model_type == 'rf':
        model = RandomForestClassifier(random_state=42)
        param_grid = {'n_estimators': [100, 200], 'max_depth': [10, 20], 'min_samples_split': [2, 5]}
    elif model_type == 'knn':
        model = KNeighborsClassifier()
        param_grid = {'n_neighbors': [3, 5, 7], 'metric': ['euclidean', 'manhattan']}
    elif model_type == 'svm':
        model = SVC(probability=True)
        param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
    elif model_type == 'gbm':
        model = GradientBoostingClassifier(random_state=42)
        param_grid = {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.1], 'max_depth': [3, 6]}

    # Perform Grid Search or RandomizedSearchCV
    if use_random_search:
        search = RandomizedSearchCV(model, param_distributions=param_grid, n_iter=10, random_state=42, cv=3)
    else:
        search = GridSearchCV(model, param_grid, cv=3)
    
    search.fit(X_train, y_train)
    return search.best_estimator_, search.best_params_

# Function to evaluate the model with ROC, confusion matrix, and KS test
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    
    # Classification report
    print("Classification Report:\n", classification_report(y_test, y_pred))
    
    # Confusion Matrix
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    
    # ROC AUC Score
    print("ROC AUC Score:", roc_auc_score(y_test, y_prob))
    
    # KS Table
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    ks_stat = max(tpr - fpr)
    print("KS Statistic:", ks_stat)

    # Plot ROC Curve
    plt.figure(figsize=(10,6))
    plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc_score(y_test, y_prob):.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.show()

    return ks_stat

def main(file_path, use_random_search=False,  models = ['logistic', 'rf', 'knn', 'svm', 'gbm'], cols=None):
    # Load and preprocess data
    X, y = load_and_preprocess_data(file_path, cols=cols)
    
    # Split data into train and test
    X_train, X_test, y_train, y_test = split_data(X, y)
        
    # Train and tune models
    best_models = {}
    for model_type in models:
        print(f"Training {model_type} model...")
        best_model, best_params = train_and_tune_model(X_train, y_train, model_type=model_type, use_random_search=use_random_search)
        print(f"Best Params for {model_type}: {best_params}")
        best_models[model_type] = best_model
    
    # Evaluate models
    ks_stats = {}
    for model_type, model in best_models.items():
        print(f"\nEvaluating {model_type} model:")
        ks_stat = evaluate_model(model, X_test, y_test)
        ks_stats[model_type] = ks_stat
    
    # Return the KS stats for comparison
    return ks_stats
```

## Model Building -- With SMOTE

```python
# Using SMOTE for imbalanced dataset
features_to_keep = ['delq_m_2', 'delinquency_severity_index', 'delinquency_acceleration', 'delinquency_persistence', 'delinquency_trend', 'months_since_last_delinquency']
from imblearn.over_sampling import SMOTE
def split_data_with_smote(X, y, test_size=0.3, random_state=42):
    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X, y)
    X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=test_size, random_state=random_state)
    print("Resampled Target Distribution:\n", y_resampled.value_counts())
    return X_train, X_test, y_train, y_test
```

```python
def main_with_smote(file_path, use_random_search=False,  models = ['logistic', 'rf', 'knn', 'svm', 'gbm'], cols=None):
    # Load and preprocess data
    X, y = load_and_preprocess_data(file_path, cols=cols)
    
    # Split data into train and test
    X_train, X_test, y_train, y_test = split_data_with_smote(X, y)
        
    # Train and tune models
    best_models = {}
    for model_type in models:
        print(f"Training {model_type} model...")
        best_model, best_params = train_and_tune_model(X_train, y_train, model_type=model_type, use_random_search=use_random_search)
        print(f"Best Params for {model_type}: {best_params}")
        best_models[model_type] = best_model
    
    # Evaluate models
    ks_stats = {}
    for model_type, model in best_models.items():
        print(f"\nEvaluating {model_type} model:")
        ks_stat = evaluate_model(model, X_test, y_test)
        ks_stats[model_type] = ks_stat
    
    # Return the KS stats for comparison
    return ks_stats
```
