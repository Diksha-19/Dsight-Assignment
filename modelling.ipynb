```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
# Separate features and target
X = data.drop(columns=['Id', 'Target'])
y = data['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Logistic Regression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
```

```python
# Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Train XGBoost model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train, y_train)

# Evaluate model
y_pred = xgb_model.predict(X_test)

# Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
models = {'Logistic Regression': log_model, 'Random Forest': rf_model, 'XGBoost': xgb_model}
```

```python
for name, model in models.items():
    print(f"\n{name} Evaluation:")
    y_pred = model.predict(X_test)
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("ROC AUC Score:", roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]))
```

### Observations from Prediction Results:

1. **Logistic Regression Performance**:
   - The model shows high accuracy of **91%**, largely due to its ability to classify the majority class (0) correctly.
   - However, the model completely fails to identify the minority class (1), with precision, recall, and F1-score all at **0.00** for class 1.
   - The **ROC AUC score** of **0.448** indicates the model is slightly worse than a random classifier for distinguishing between classes.

2. **Random Forest Performance**:
   - The model achieves an accuracy of **89%**, but like Logistic Regression, it struggles with class imbalance, completely failing to predict the minority class (1).
   - Precision, recall, and F1-score for class 1 are **0.00**, and the **ROC AUC score** is **0.397**, highlighting poor overall performance in class separation.

3. **XGBoost Performance**:
   - XGBoost achieves an accuracy of **88%**, slightly lower than the other models.
   - Similar to Logistic Regression and Random Forest, it fails to predict the minority class (1), resulting in **0.00** precision, recall, and F1-score for this class.
   - The **ROC AUC score** of **0.429** indicates poor discrimination ability for the minority class.

4. **Common Trends Across Models**:
   - All models exhibit strong bias towards the majority class (0), likely due to the severe class imbalance in the dataset.
   - Precision, recall, and F1-score for the minority class (1) are consistently **0.00** across all models.
   - ROC AUC scores for all models are significantly below 0.5, reflecting poor performance in distinguishing between the classes.

Let's handle class imbalance and retrain the models to see if the performance improves.

```python
# Using SMOTE for imbalanced dataset

from imblearn.over_sampling import SMOTE

# Separate features and target
X = data.drop(columns=['Id', 'Target'])
y = data['Target']

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Check new target distribution
print("Resampled Target Distribution:\n", y_resampled.value_counts())
```

```python
# Regenerate train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)
```

```python
# Logistic Regression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
```

```python
# Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Train XGBoost model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train, y_train)

# Evaluate model
y_pred = xgb_model.predict(X_test)

# Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
smote_models = {'Logistic Regression': log_model, 'Random Forest': rf_model, 'XGBoost': xgb_model}
```

```python
# Evaluate each model
models = {'Logistic Regression': log_model, 'Random Forest': rf_model, 'XGBoost': xgb_model}
for name, model in models.items():
    print(f"\n{name} Evaluation:")
    y_pred = model.predict(X_test)
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("ROC AUC Score:", roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]))
```

### Observations from Prediction Results:

#### **Logistic Regression**:
1. **Moderate Performance**:
   - Accuracy is **61%**, indicating moderate performance on the dataset.
   - The **ROC AUC score** of **0.67** suggests that the model has some ability to differentiate between the classes, but it is not highly effective.

2. **Imbalanced Precision-Recall**:
   - Class `0` (Non-default) has **higher precision (0.65)** but lower recall (**0.53**), meaning many non-default cases are being misclassified as default.
   - Class `1` (Default) shows the opposite pattern with **lower precision (0.58)** and higher recall (**0.69**), meaning the model predicts more defaults but with less accuracy.

3. **Room for Improvement**:
   - Logistic Regression struggles with non-linear relationships and imbalanced datasets, which may explain its limited performance.

2. **Balanced Performance**:
   - Both classes (`0` and `1`) show strong precision (**0.90, 0.89**) and recall (**0.89, 0.90**).
   - The confusion matrix shows that very few samples are misclassified, suggesting good generalization.

3. **Strength of Random Forest**:

#### **XGBoost**:
